## Create new Anaconda (virtual) environment  (Optional) 
In your conda environment create a new environment.

```sh
conda create --name venv
 ```
 Where `venv` is your virtual environment name.

### Activate environment
 
 ```sh
conda activate venv
 ```
 You will need to install all packages in the new virtual enviroment, since it will be created empty.
 
 
 #### Install nb_conda
 This is recommended if you want to keep switching between kernels through a graphical interface.
 
 ```sh
(venv)conda install nb_conda
 ```
 
 You will need to deactivate and activate your enviroment so the previous instruction takes place. 
 

### Dectivate environment
 Everytime you install a new package you need to deactivate and (re) activate your environment.
   
 ```sh
conda deactivate
 ```
In the `Kernel` option, you will see the environments you can use in your environment.

* This short [video](https://www.youtube.com/watch?v=Ro9l0eapoJU) shows these steps in more details. 
 
* The creation of real virtual environment is slightly different. Look [here](https://medium.com/@rahul3012_37725/a-very-basic-guide-to-python-virtual-environments-a53d1e191490) for more details.

* More detailed documentation about [venv](https://docs.python.org/3/tutorial/venv.html).




### Using Jupyter Notebooks

1. Open up the Anaconda/Miniconda terminal
2. Navigate to the folder containing the notebook
3. Activate the `venv` environment you created
4. Start up the notebook

```sh
source activate venv
jupyter notebook
```
